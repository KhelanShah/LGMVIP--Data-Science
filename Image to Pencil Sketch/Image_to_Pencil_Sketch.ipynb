### **IMAGE TO PENCIL SKETCH**

**Step:1 The first step is to import the libraries that are required for the implementation of the code operation.Here we import google colab patches to import an image.**

```python
from google.colab.patches import cv2_imshow
import cv2
```

**Step:2 Importing an image.**

```python
img = cv2.imread('dog.jpg')
cv2_imshow(img)
```

**Step:3 Converting an image into gray scale.**

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(img_gray)
```

**Step:4 Inverting an image.**

```python
img_invert = cv2.bitwise_not(img_gray)
cv2_imshow(img_invert)
```

**Step:5 Smoothing an image.**

```python
img_smoothing = cv2.GaussianBlur(img_invert, (21, 21),sigmaX=0, sigmaY=0)
cv2_imshow(img_smoothing)
```

**Step:6 Obtaining an final image.**

```python
def dodgeV2(x, y):
    return cv2.divide(x, 255 - y, scale=256)
```

```python
final_img = dodgeV2(img_gray, img_smoothing)
cv2_imshow(final_img)
```
